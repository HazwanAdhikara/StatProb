## SOAL 1

Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

```python
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
```

### RANGE

```python
# Fungsi untuk menghitung rentang
def calculate_range(data):
    nilai_tertinggi = max(data)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(data)
    print(f"Nilai terendah: {nilai_terendah}")

    # Selisih antara nilai tertinggi dan terendah
    data_range = max(data) - min(data)
     
    print(f"\n")  
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  

    return data_range


# Print dataset
  
print(f"Data: {data}")

data_range = calculate_range(data)
```

#### Visualisasi

```python
# Membuat Histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Membuat batas koordinat
y = 1 # Koordinat Y dimana garis akan digambar
x1 = 4/70*1 # X mula-mula
x2 = 66/70*1 # X akhir

# Garis Horizontal
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(data)}')

# Garis Vertikal
plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=max(data), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')

# Membuat Label
plt.xlabel('Jumlah Gol')
plt.ylabel('Frekuensi')
plt.title('Range')

# Menambah Legenda
plt.legend()

# Menunjukkan Plot
plt.show()
```

### VARIANSI

```python
# Fungsi untuk menghitung rata-rata
def calculate_mean(data):
    return sum(data) / len(data)

# Fungsi untuk menghitung variansi
def calculate_variance(data):
    # Rata-rata 
    mean = calculate_mean(data)
    print(f"Rata-rata: {mean}")  
    
    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")    
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")    
    
    print(f"Total kuadrat selisih: {sum(squared_diff)}")    
    print(f"Jumlah total data: {len(data)}")    
    
    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(data)
    
    print(f"\n")    
    print(f"Variansi: {sum(squared_diff)} / {len(data)} = {variance}")    
    
    return variance


# Print Dataset
print(f"Data: {data}")

variansi = calculate_variance(data)
```

### STANDAR DEVIASI

```python
# Fungsi untuk menghitung standar deviasi
def calculate_standard_deviation(data):
    variansi = calculate_variance(data)
    standar_deviasi = variansi**0.5
    print(f"\n")   
    print(f"Standard Deviation: {variansi}**0.5 = {standar_deviasi}")
    
    return standar_deviasi

# Print dataset
print(f"Data: {data}")

standar_deviasi = calculate_standard_deviation(data)
mean = calculate_mean(data)
```

#### Visualisasi

```python
# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standar_deviasi, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standar_deviasi:.2f}')
plt.axvline(x=mean + standar_deviasi, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standar_deviasi:.2f}')

# Tambahkan label dan judul
plt.xlabel('Jumlah Gol')
plt.ylabel('Frekuensi')
plt.title(f'Standar Deviasi = {standar_deviasi:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```

dari sini kita tau bahwa, ada OUTLIER yaitu 9, 10, 17, dan 19. Oleh karena itu, kita perlu menginisialisasi dataset yang diperbaharui

#### Data yang baru

```python
databaru = [12, 13, 15, 14, 13]
```

```python
print(f"Data: {databaru}")

standar_deviasi = calculate_standard_deviation(databaru)
mean = calculate_mean(databaru)
```

### KESIMPULAN

Kesimpulannya, Standar Deviasi yang besar memberikan variansi data yang lebih besar dan dapat memberian informasi yang lebih berharga. Sedangkan, dengan Standar Deviasi yang lebih kecil dapat memberikan tingkat konsistensi atau prediktabilitas yang lebih tinggi dalam dataset.
