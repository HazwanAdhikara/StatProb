```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

## SOAL 1 

Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

```python
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
```

### MEAN

#### Manual

```python
print(f"Data: {data}")

# Jumlah nilai dalam dataset
total_nilai = sum(data)
print(f"Total Nilai: {total_nilai}")

# banyaknya data dalam dataset
banyak_data = len(data)
print(f"Banyak Data: {banyak_data}")

# Rata-rata dari data
mean_manual = total_nilai / banyak_data
print(f"Mean: {mean_manual}")
```

#### Numpy

```python
mean_data = np.mean(data)
print(f"Mean = {mean_data}")
```

#### Visualisasi

```python
# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='yellow')

# Garis penanda mean
plt.axvline(x = mean_manual, color ='red', linestyle ='solid', linewidth = 2, label = f'Mean = {mean_manual:.2f}')
plt.axvline(x = mean_data, color ='black', linestyle ='dashed', linewidth = 1, label = f'Mean = {mean_data:.2f}')

# Menambah Label
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Mean')

# Menambah Legend
plt.legend()

# Menunjukkan Plot
plt.show()
```

saya membuat 2 garis untuk penanda garis mean, dengan metode manual dan numpy, dan ternyata rata rata yang dihasilkan sama

### MEDIAN

#### Manual

```python
print(f"Data: {data}")

# Mengurutkan data dari terkecil hingga terbesar
urutan_data = sorted(data)
print(f"Data berurut: {urutan_data}")

# Banyak data dalam dataset 
n = len(urutan_data)
print(f"Banyaknya Data: {n}")

nilai_tengah = n//2 # digunakan untuk mencari titik tengah

# Jika jumlah data genap, median adalah rata-rata dari dua nilai tengah 
if n % 2 == 0:
    print(f"Jumlah total data Genap")
    median1 = urutan_data[nilai_tengah]
    median2 = urutan_data[nilai_tengah - 1]
    print(f"Nilai tengah: {median1} {median2}")
    median_manual = (median1 + median2) / 2
    
# Jika jumlah data ganjil, median adalah nilai tengah
else:
    print(f"Jumlah total data Ganjil")
    median_manual = urutan_data[nilai_tengah]
    print(f"Nilai tengah: {median_manual}")

print(f"Median: {median_manual}")
```

#### Numpy

```python
median_data = np.median(data)
print(f"Median = {median_data}")
```

#### Visualisasi

```python
# Membuat Histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='blue')

# Penanda dengan Garis Lurus
plt.axvline(x=median_manual, color='yellow', linestyle='solid', linewidth=2, label=f'Median = {median_manual:.2f}')
plt.axvline(x=median_data, color='black', linestyle='dashed', linewidth=1, label=f'Median = {median_data:.2f}')

# Menambah Label
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Median')

# Menambah legend
plt.legend()

# Menunjukkan Plot
plt.show()
```

saya membuat 2 garis untuk penanda garis Median, dengan metode manual dan numpy, dan ternyata Median yang dihasilkan sama

### MODUS

#### Manual

```python
print(f"Data: {data}")

penghitung = Counter(data)
mode_data = dict(penghitung) # ubah ke tipe data dictionary
max_penghitung = max(list(penghitung.values())) # mencari data yang memiliki kemunculan paling banyak
print(f"Max Kemunculan: {max_penghitung}")

modus_data = []
for k, v in mode_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(penghitung.values())):
        modus_data.append(k)
    
print(f"Modus: {modus_data}")
```

#### Pake Library

```python
try:
    modus_lib = stats.mode(data).mode[0]
except IndexError:
    modus_lib = "Tidak ada Modus"

print(f'Modus: {modus_lib}')
```

#### Visualisasi

```python
# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Membuat garis lurus pada modus
for k in modus_data:
    plt.axvline(x=k, color='red', linestyle='dashed', linewidth=1, label=f'Mode = {k:.2f}')

# Menambah Label
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Modus')

# Menambah Legend
plt.legend()

# Menunjukkan Plot
plt.show()
```

karena dalam data tidak ada modusnya, maka semua data yang ada dalam dataset tersebut adalah modusnya, karena semua data tersebut muncul paling banyak 1 kali
